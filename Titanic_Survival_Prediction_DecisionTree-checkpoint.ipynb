```python
#Data Cleaning
#Drop the name column
train_df.drop('Name',axis=1,inplace=True)
test_df.drop('Name',axis=1,inplace=True)
#Drop the ticket column
train_df.drop('Ticket',axis=1,inplace=True)
test_df.drop('Ticket',axis=1,inplace=True)
train_df.head()
```

```python
#To convert the categorical data in Sex column by mapping 1 :Male and 0:Female#Do it for both test and train data
for gender in train_df['Sex']:
    if gender=='male':
        train_df['Sex']=train_df['Sex'].replace(to_replace='male',value=1)
    else:
        train_df['Sex']=train_df['Sex'].replace(to_replace='female',value=0)
#Converting in the test data
for gender in test_df['Sex']:
    if gender=='male':
        test_df['Sex']=test_df['Sex'].replace(to_replace='male',value=1)
    else:
        test_df['Sex']=test_df['Sex'].replace(to_replace='female',value=0)
train_df.dtypes
```

```python
#Data cleaning with the Cabin
#data cleaning for Cabin variable,since 687 records have not recorded their values
train_df['Cabin'].fillna('G6',inplace=True)
import _string
CabinToNum=0
for x in train_df['Cabin']:
    for letter in x[0:1]:
            cabin_letter=letter.strip()
    train_df['Cabin']=train_df['Cabin'].replace(to_replace=x,value=cabin_letter)#Replace with only the cabin names:A to G
for cabin_name in train_df['Cabin']:
    if (cabin_name=='A'):
        train_df['Cabin']=train_df['Cabin'].replace(to_replace='A',value=1)
    elif (cabin_name=='B'):
        train_df['Cabin']=train_df['Cabin'].replace(to_replace='B',value=2)
    elif (cabin_name=='C'):
        train_df['Cabin']=train_df['Cabin'].replace(to_replace='C',value=3)
    elif (cabin_name=='D'):
        train_df['Cabin']=train_df['Cabin'].replace(to_replace='D',value=4)
    elif (cabin_name=='E'):
        train_df['Cabin']=train_df['Cabin'].replace(to_replace='E',value=5)
    elif (cabin_name=='F'):
        train_df['Cabin']=train_df['Cabin'].replace(to_replace='F',value=6)
    else:
        train_df['Cabin']=train_df['Cabin'].replace(to_replace='G',value=7)

 #Data cleaning for test data
test_df['Cabin'].fillna('G6',inplace=True)
import _string
CabinToNum=0
for x in test_df['Cabin']:
    for letter in x[0:1]:
            cabin_letter=letter.strip()
    test_df['Cabin']=test_df['Cabin'].replace(to_replace=x,value=cabin_letter)#Replace with only the cabin names:A to G
for cabin_name in test_df['Cabin']:
    if (cabin_name=='A'):
        test_df['Cabin']=test_df['Cabin'].replace(to_replace='A',value=0)
    elif (cabin_name=='B'):
        test_df['Cabin']=test_df['Cabin'].replace(to_replace='B',value=1)
    elif (cabin_name=='C'):
        test_df['Cabin']=test_df['Cabin'].replace(to_replace='C',value=2)
    elif (cabin_name=='D'):
        test_df['Cabin']=test_df['Cabin'].replace(to_replace='D',value=3)
    elif (cabin_name=='E'):
        test_df['Cabin']=test_df['Cabin'].replace(to_replace='E',value=4)
    elif (cabin_name=='F'):
        test_df['Cabin']=test_df['Cabin'].replace(to_replace='F',value=5)
    else:
        test_df['Cabin']=test_df['Cabin'].replace(to_replace='G',value=6) 
#The cabin column contains a value as T,so it will return value error on tring to fit the model.So lets drop that row
train_df.drop(train_df.loc[train_df['Cabin']=='T'].index, inplace=True)
train_df.drop(train_df.loc[train_df['Cabin']=='C'].index, inplace=True)
```

```python
train_df.head()
```

```python
#Data cleaning with age group in the training set
train_df['Age'].fillna(train_df['Age'].mean(),inplace=True)
test_df['Age'].fillna(test_df['Age'].mean(),inplace=True)
for age in train_df['Age']:
    if (age < 10):
        train_df['Age']=train_df['Age'].replace(to_replace=age,value='A')
    elif (age>=10) and (age<=25):
        train_df['Age']=train_df['Age'].replace(to_replace=age,value='B')  
    elif (age>25) and (age<=46):
        train_df['Age']=train_df['Age'].replace(to_replace=age,value='C')
    elif (age>46) and (age<=60):
        train_df['Age']=train_df['Age'].replace(to_replace=age,value='D')
    else:
        train_df['Age']=train_df['Age'].replace(to_replace=age,value='E')
#Data cleaning with the test data
for age in test_df['Age']:
    if (age < 10):
        test_df['Age']=test_df['Age'].replace(to_replace=age,value='A')
    elif (age>=10) and (age<=25):
        test_df['Age']=test_df['Age'].replace(to_replace=age,value='B')  
    elif (age>25) and (age<=46):
        test_df['Age']=test_df['Age'].replace(to_replace=age,value='C')
    elif (age>46) and (age<=60):
        test_df['Age']=test_df['Age'].replace(to_replace=age,value='D')
    else:
        test_df['Age']=test_df['Age'].replace(to_replace=age,value='E')
```

```python
for age in train_df['Age']:
    if (age=='A'):
        train_df['Age']=train_df['Age'].replace(to_replace='A',value=0.0)
    elif (age=='B'):
        train_df['Age']=train_df['Age'].replace(to_replace='B',value=1.0)
    elif (age=='C'):
        train_df['Age']=train_df['Age'].replace(to_replace='C',value=2.0)
    elif (age=='D'):
        train_df['Age']=train_df['Age'].replace(to_replace='D',value=3.0)
    else:
        train_df['Age']=train_df['Age'].replace(to_replace='E',value=4)
#Testing data
for age in test_df['Age']:
    if (age=='A'):
        test_df['Age']=test_df['Age'].replace(to_replace='A',value=0)
    elif (age=='B'):
        test_df['Age']=test_df['Age'].replace(to_replace='B',value=1)
    elif (age=='C'):
        test_df['Age']=test_df['Age'].replace(to_replace='C',value=2)
    elif (age=='D'):
        test_df['Age']=test_df['Age'].replace(to_replace='D',value=3)
    else:
        test_df['Age']=test_df['Age'].replace(to_replace='E',value=4)
train_df.dtypes
```

```python

train_df.dtypes
```

```python
#Data Cleaning with Embarked Column
train_df['Embarked'].fillna('S',inplace=True)
for embarked in train_df['Embarked']:
    if (embarked=='S'):
        train_df['Embarked']=train_df['Embarked'].replace(to_replace='S',value=0)
    elif (embarked=='C'):
        train_df['Embarked']=train_df['Embarked'].replace(to_replace='C',value=1)
    else:
        train_df['Embarked']=train_df['Embarked'].replace(to_replace='Q',value=2)
##Data Cleaning with Embarked Column in the testing data
test_df['Embarked'].fillna('S',inplace=True)
for embarked in test_df['Embarked']:
    if (embarked=='S'):
        test_df['Embarked']=test_df['Embarked'].replace(to_replace='S',value=0)
    elif (embarked=='C'):
        test_df['Embarked']=test_df['Embarked'].replace(to_replace='C',value=1)
    else:
        test_df['Embarked']=test_df['Embarked'].replace(to_replace='Q',value=2)
train_df.dtypes
```

```python

X=pd.DataFrame(np.c_[train_df['PassengerId'],train_df['Pclass'],train_df['Sex'],train_df['Age'],train_df['SibSp'],train_df['Parch'],train_df['Cabin'],train_df['Embarked']],columns=['PassengerId','Pclass','Sex','Age','SibSp','Parch','Cabin','Embarked'])
Y=train_df['Survived']
```

```python
#Building the model
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=.25,random_state=9)
print(Y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
reg_model=DecisionTreeClassifier(criterion="entropy",max_depth=10,min_samples_split=10)
reg_model.fit(X_train,Y_train)
```

```python
predictions=reg_model.predict(X_test)
print(predictions[0:10])
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(Y_test,predictions)
#print(predictions.shape)
#print(Y_test)
print(score)
```
